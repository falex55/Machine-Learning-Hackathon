```python
# speration in independant and dependant features

X = df2.drop("Birth_Weight",axis=1) 
y = df2["Birth_Weight"]
```

```python
# separate dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=0)

X_train.shape, X_test.shape
```

```python
# correlation of traning data 

import seaborn as sns
#Using Pearson Correlation
plt.figure(figsize=(12,10))
cor = X_train.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
# pipeline for scaling and also null values imputation , there is no any missing data in this dataset but still i have implement
# this for future data

from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    #     ..... add as many as you want in your pipeline
    ('std_scaler', StandardScaler()),
])
```

```python
X= my_pipeline.fit_transform(X)
```

```python
print(X.shape)
X = pd.DataFrame(X)
X.head()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import r2_score
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

## Implementation of KNN algorithm

```python
# function for finding best neighbor value / hyperparameter tunning
error_rate = []
for i in range(1,40):
    
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(X_train,y_train)
    predctions_i = knn.predict(X_test)
    error_rate.append(np.mean(predctions_i != y_test))
```

```python
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn_Regressor=KNeighborsRegressor(n_neighbors=1)
knn_Regressor.fit(X_train,y_train)
knn_predictions=knn_Regressor.predict(X_test)
knn_mean_squared_error=mean_squared_error(y_test,knn_predictions)
print('mse : ',knn_mean_squared_error)
rmse = math.sqrt(knn_mean_squared_error)
print('rmse: ', rmse)
print('R2 score: ',r2_score(y_test, knn_predictions))
```

## Implementation of random Forest using GridSearchCV algorithm

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
```

```python

from sklearn.model_selection import GridSearchCV

# Create the parameter grid based on the results of random search 
param_grid = {
    'bootstrap': [True],
    'max_depth': [10,15],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5,6],
    'min_samples_split': [3,4,5,6],
    'n_estimators': [100, 200, 250, 300, 350, 400]
}
# Create a based model
rf = RandomForestRegressor()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 0)

# Fit the grid search to the data
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
best_grid = grid_search.best_estimator_
best_grid
```

```python
rf_predict = grid_search.predict(X_test)
```

```python
rf_mean_squared_error=mean_squared_error(y_test,rf_predict)
print('mse : ',rf_mean_squared_error)
rmse = math.sqrt(rf_mean_squared_error)
print('rmse: ', rmse)
print('R2 score: ',r2_score(y_test, rf_predict))
```
